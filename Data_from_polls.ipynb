```python
# Read pdf into DataFrame
#https://stackoverflow.com/questions/45457054/tabula-extract-tables-by-area-coordinates
#pobieranie danych dla diecezji
#https://docs.google.com/spreadsheets/d/1WHCrTZX2WJeYStcG58wh6y9Q58bwhGRG92iLxz4Paek/edit#gid=752560630
df = tabula.read_pdf("dane_pdf/analiza_elektoratu/2019/K_038_17.PDF",
                     multiple_tables=True, 
                     pages=[9],
                     #stream = True
                     guess = True,  
                     #pandas_options={'header':True},
                     #area=(20, 40, 695, 589),
                     spreadsheet=True
                    )
name = '2017_04'
```
