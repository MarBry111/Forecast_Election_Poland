```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from os import listdir
```

```python
files = listdir('model/model1')
```

```python
output = []
with open('model/model1/'+files[3], "r") as f:
    for fline in f.read().split('\n')[:-1]:
        fline = fline[3:].replace('[','').split(']')
        tmp_output = []
        for fl in fline:
            if(len(fl)>1):
                tmptmp_out = []
                for fi in fl.split(' '):
                    if(len(fi)>1): tmptmp_out.append(float(fi))
                tmp_output.append(tmptmp_out)
        output.append(tmp_output)
```
