# Dane - pytanie elektoratów

```python
df = tabula.read_pdf("dane_pdf/charaktetyryka_pogladow/pdf/K_124_19.PDF",
                     multiple_tables=True, 
                     pages=[21,22],
                     #stream = True
                     guess = True,  
                     #pandas_options={'header':True},
                     #area=(20, 40, 695, 589),
                     spreadsheet=True
                    )
```

```python
df_names = df[0]
```

## Points

```python
for i in range(len(df)):
    df[i] = df[i].dropna().reset_index()
    df[i]
catg = 'points'
```

## Percent

```python
for i in range(len(df)):
    df[i] = df[i].drop(0)
    df[i] = df[i].drop(2)
    df[i]
catg = 'percent'
```

```python
result = pd.concat([df_names,df[0],df[1]]).reset_index()
del result['index']
try:
    del result['level_0']
except:
    pass
result
```

```python
result = result.drop_duplicates(subset=None, keep='first', inplace=False)
```

```python
result = result.apply(lambda x: x.str.replace(',','.'))
```

```python
result[0] = [q.replace('\n',' ').replace('\r','') if type(q) == str else 'Question' for q in result[0]]
result.loc[0] = [q.replace('\n',' ').replace('\r','') if type(q) == str else 'Question' for q in result.loc[0]]
```

```python
result
```

```python
result.to_csv('dane_pdf/charaktetyryka_pogladow/2019_'+catg+'.csv', index = False, header=False)
```
